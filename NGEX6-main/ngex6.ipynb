Some typical NumPy 'random' functions:

```python
import numpy as np

# To generate a random float values between 0 and 1 (exclusive)
print(np.random.rand(3, 2)) # Creates a 3 x 2 matrix of random float values between 0 and 1 (exclusive)

# To generate random float values from standard normal distribution (mean = 0 and std = 1)
print(np.random.randn(5)) # Generates 5 random float values from standard normal distribution

# To generate random integer values within a range of values
print(np.random.randint(1, 100, 10)) # Generates 10 random integer values between 1 and 100 (exclusive)

# To randomly select an element from a given list of elements
print(np.random.choice([1, 2, 3, 4, 5, 6, 7, 8, 9], size=3)) # Three random numbers from the given array will be chosen

# To generate a random sample of values from a normally distributed with a specified mean and standard deviation
print(np.random.normal(loc=0, scale=1, size=10)) # 10 random values will be generated from a normally distributed data where 0 (loc) is the mean and 1 (scale) is the standard deviation

# To set a seed for a random number generation to ensure you get the same results later as well
print(np.random.seed(42)) # This ensures that any random numbers generated in the program are same every time you run the program. The seed value '42' can be any other value
```

```python
import numpy as np


data = np.genfromtxt('sleep_health.csv', delimiter=',', skip_header=1, dtype=None, encoding='utf-8', names=True)

print("Column Names:", data.dtype.names)

age = data['Age'] 
sleep_duration = data['Sleep_Duration']  
heart_rate = data['Heart_Rate'] 
daily_steps = data['Daily_Steps'] 



corr_sleep_age = np.corrcoef(sleep_duration, age)[0, 1]

corr_sleep_heart = np.corrcoef(sleep_duration, heart_rate)[0, 1]


corr_sleep_steps = np.corrcoef(sleep_duration, daily_steps)[0, 1]


std_sleep_duration = np.std(sleep_duration)


print(f"Correlation between Sleep Duration and Age: {corr_sleep_age}")
print(f"Correlation between Sleep Duration and Heart Rate: {corr_sleep_heart}")
print(f"Correlation between Sleep Duration and Daily Steps: {corr_sleep_steps}")
print(f"Standard Deviation of Sleep Duration: {std_sleep_duration}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time


df = pd.DataFrame({
    'Age': [25, 30, 22, 35, 40, 29, 31],
    'Sleep Duration': [7, 8, 6, 5, 7.5, 6.5, 8],
    'Quality of Sleep': [3, 4, 2, 5, 4, 3, 5],
    'Physical Activity Level': [3, 2, 4, 5, 3, 2, 4],
    'Stress Level': [2, 3, 4, 3, 2, 1, 2],
    'Heart Rate': [70, 75, 65, 80, 78, 72, 71],
    'Occupation': ['Student', 'Worker', 'Student', 'Worker', 'Worker', 'Student', 'Worker'],
    'BMI': [22, 25, 21, 28, 26, 24, 23],
    'Daily Steps': [5000, 8000, 7000, 10000, 9000, 6000, 7500]
})


def plot_distributions_with_pandas(df):
    distributions = [
        ('Age', 'Age Distribution'),
        ('Sleep Duration', 'Sleep Duration Distribution'),
        ('Quality of Sleep', 'Quality of Sleep Distribution'),
        ('Physical Activity Level', 'Physical Activity Level Distribution'),
        ('Stress Level', 'Stress Level Distribution'),
        ('Heart Rate', 'Heart Rate Distribution'),
    ]

    for column, title in distributions:
        plt.figure(figsize=(10, 6))
        sns.histplot(df[column], bins=10, kde=True)
        plt.title(title)
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.show()


def plot_distributions_with_numpy(df):
    distributions = [
        ('Age', 'Age Distribution'),
        ('Sleep Duration', 'Sleep Duration Distribution'),
        ('Quality of Sleep', 'Quality of Sleep Distribution'),
        ('Physical Activity Level', 'Physical Activity Level Distribution'),
        ('Stress Level', 'Stress Level Distribution'),
        ('Heart Rate', 'Heart Rate Distribution'),
    ]

    for column, title in distributions:
        plt.figure(figsize=(10, 6))
        plt.hist(df[column], bins=10, density=True, alpha=0.6)
        plt.title(title)
        plt.xlabel(column)
        plt.ylabel('Density')
        plt.show()


def plot_distributions_by_factor_with_pandas(df):
    factors = [
        ('Quality of Sleep', 'Sleep Duration by Quality of Sleep'),
        ('Stress Level', 'Sleep Duration by Stress Level'),
        ('Physical Activity Level', 'Sleep Duration by Physical Activity Level'),
        ('Occupation', 'Sleep Duration by Occupation'),
        ('BMI', 'Sleep Duration by BMI'),
    ]

    for factor, title in factors:
        plt.figure(figsize=(10, 6))
        sns.boxplot(x=factor, y='Sleep Duration', data=df)
        plt.title(title)
        plt.xlabel(factor)
        plt.ylabel('Sleep Duration (hours)')
        plt.xticks(rotation=45)
        plt.show()


def plot_distributions_by_factor_with_numpy(df):
    factors = [
        ('Quality of Sleep', 'Sleep Duration by Quality of Sleep'),
        ('Stress Level', 'Sleep Duration by Stress Level'),
        ('Physical Activity Level', 'Sleep Duration by Physical Activity Level'),
        ('Occupation', 'Sleep Duration by Occupation'),
    ]

    for factor, title in factors:
        plt.figure(figsize=(10, 6))
        for category in np.unique(df[factor]):
            plt.hist(df['Sleep Duration'][df[factor] == category], bins=10, alpha=0.5, label=f'{factor} {category}')
        plt.title(title)
        plt.xlabel('Sleep Duration (hours)')
        plt.ylabel('Frequency')
        plt.legend()
        plt.show()


def plot_relationships_with_pandas(df):
    relationships = [
        ('Age', 'Sleep Duration', 'Age vs Sleep Duration'),
        ('Sleep Duration', 'Heart Rate', 'Sleep Duration vs Heart Rate'),
        ('Heart Rate', 'Daily Steps', 'Heart Rate vs Daily Steps'),
        ('Sleep Duration', 'Daily Steps', 'Sleep Duration vs Daily Steps'),
    ]

    for x, y, title in relationships:
        plt.figure(figsize=(10, 6))
        sns.scatterplot(x=x, y=y, data=df)
        plt.title(title)
        plt.xlabel(x)
        plt.ylabel(y)
        plt.show()


def plot_relationships_with_numpy(df):
    relationships = [
        ('Age', 'Sleep Duration', 'Age vs Sleep Duration'),
        ('Sleep Duration', 'Heart Rate', 'Sleep Duration vs Heart Rate'),
        ('Heart Rate', 'Daily Steps', 'Heart Rate vs Daily Steps'),
        ('Sleep Duration', 'Daily Steps', 'Sleep Duration vs Daily Steps'),
    ]

    for x, y, title in relationships:
        plt.figure(figsize=(10, 6))
        plt.scatter(df[x], df[y])
        plt.title(title)
        plt.xlabel(x)
        plt.ylabel(y)
        plt.show()


start_time_pandas = time.time()
plot_distributions_with_pandas(df)
plot_distributions_by_factor_with_pandas(df)
plot_relationships_with_pandas(df)
end_time_pandas = time.time()

start_time_numpy = time.time()
plot_distributions_with_numpy(df)
plot_distributions_by_factor_with_numpy(df)
plot_relationships_with_numpy(df)
end_time_numpy = time.time()


pandas_time = end_time_pandas - start_time_pandas
numpy_time = end_time_numpy - start_time_numpy

print(f'Time taken using Pandas: {pandas_time:.4f} seconds')
print(f'Time taken using NumPy: {numpy_time:.4f} seconds')
```

AIM #4: Other possible plotting

1. Think of other possible plots to show some interesting distribution and relations. Do this using both pandas and NumPy



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time


df = pd.DataFrame({
    'Age': [25, 30, 22, 35, 40, 29, 31],
    'Sleep Duration': [7, 8, 6, 5, 7.5, 6.5, 8],
    'Quality of Sleep': [3, 4, 2, 5, 4, 3, 5],
    'Physical Activity Level': [3, 2, 4, 5, 3, 2, 4],
    'Stress Level': [2, 3, 4, 3, 2, 1, 2],
    'Heart Rate': [70, 75, 65, 80, 78, 72, 71],
    'Occupation': ['Student', 'Worker', 'Student', 'Worker', 'Worker', 'Student', 'Worker'],
    'BMI': [22, 25, 21, 28, 26, 24, 23],
    'Daily Steps': [5000, 8000, 7000, 10000, 9000, 6000, 7500]
})




def plot_pair_plot_with_pandas(df):
    plt.figure(figsize=(10, 8))
    sns.pairplot(df, diag_kind='kde')
    plt.suptitle('Pair Plot of Variables', y=1.02)
    plt.show()


def plot_heatmap_with_pandas(df):
    plt.figure(figsize=(10, 8))
    correlation_matrix = df.corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Heatmap')
    plt.show()


def plot_violin_plot_with_pandas(df):
    plt.figure(figsize=(10, 6))
    sns.violinplot(x='Quality of Sleep', y='Sleep Duration', data=df)
    plt.title('Violin Plot of Sleep Duration by Quality of Sleep')
    plt.xlabel('Quality of Sleep')
    plt.ylabel('Sleep Duration (hours)')
    plt.show()


def plot_facet_grid_with_pandas(df):
    g = sns.FacetGrid(df, col='Occupation', height=4)
    g.map(sns.histplot, 'Sleep Duration', kde=True)
    g.set_titles(col_template="{col_name}")
    plt.subplots_adjust(top=0.9)
    g.fig.suptitle('Distribution of Sleep Duration by Occupation')
    plt.show()


print("Pandas Additional Plots:")
plot_pair_plot_with_pandas(df)
plot_heatmap_with_pandas(df)
plot_violin_plot_with_pandas(df)
plot_facet_grid_with_pandas(df)


def plot_heatmap_with_numpy(df):
    correlation_matrix = np.corrcoef(df.select_dtypes(include=[np.number]).values, rowvar=False)
    plt.figure(figsize=(10, 8))
    plt.imshow(correlation_matrix, cmap='coolwarm', aspect='auto')
    plt.colorbar()
    plt.title('Correlation Heatmap (NumPy)')
    plt.xticks(ticks=np.arange(len(df.columns)), labels=df.columns, rotation=45)
    plt.yticks(ticks=np.arange(len(df.columns)), labels=df.columns)
    plt.show()


print("NumPy Additional Plots:")
plot_heatmap_with_numpy(df)
```
